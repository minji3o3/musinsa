## 브랜드 별 조회수 평균 계산

```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("musinsa_data.csv")

# 브랜드 별 합
df_b = df.groupby('브랜드').sum()[['총조회수']].reset_index()

# 브랜드 별 총 제품개수 확인
sorted = pd.DataFrame(df.groupby('브랜드').count()['제품명'].sort_values(ascending=False)).reset_index()
sorted['제품개수'] = sorted['제품명']

# 두개의 데이터 프레임 병합하여, 브랜드 별 제품 조회수 평균 값 구함
df_view = pd.merge(sorted, df_b, how = 'left')
df_view['브랜드_조회수_평균'] = df_view['총조회수']/df_view['제품개수']
df_view = df_view.sort_values(by='브랜드_조회수_평균', ascending=False)
df_view = df_view.reset_index()
del df_view['index']

# 조회수 결측치 확인
df_view[df_view['총조회수'] == 0]

# 데이터 프레임 저장
df_view.to_csv("brand_view.csv", encoding = 'utf-8')
```

## 브랜드 별 구매수 평균 계산

```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("musinsa_data.csv")

# 브랜드 별 합
df_b = df.groupby('브랜드').sum()[['총구매수']].reset_index()

# 브랜드 별 총 제품개수 확인
sorted = pd.DataFrame(df.groupby('브랜드').count()['제품명'].sort_values(ascending=False)).reset_index()
sorted['제품개수'] = sorted['제품명']

# 두개의 데이터 프레임 병합하여, 브랜드 별 제품 구매수 평균 값 구함
df_purchase = pd.merge(sorted, df_b, how = 'left')
df_purchase['브랜드_구매수_평균'] = df_purchase['총구매수']/df_purchase['제품개수']
df_purchase = df_purchase.sort_values(by='브랜드_구매수_평균', ascending=False)
df_purchase = df_purchase.reset_index()
del df_purchase['index']

# 구매수 결측치 확인
df_purchase[df_purchase['총구매수'] == 0]

# 데이터 프레임 저장
df_purchase.to_csv("brand_purchase.csv", encoding = 'utf-8')
```
