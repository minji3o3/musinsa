```python
import os
import cv2
import numpy as np
import requests
from PIL import Image
from io import BytesIO
from ultralytics import YOLO

#의류 판별 모델
# YOLO 모델 파일 및 설정 파일 경로
weights_path = 'yolov3.weights'
config_path = 'yolov3.cfg'
names_path = 'coco.names'

# YOLO 네트워크 로드
net = cv2.dnn.readNet(weights_path, config_path)

# 클래스 이름 로드
with open(names_path, 'r') as f:
    classes = [line.strip() for line in f.readlines()]

# 사람 클래스의 인덱스 찾기
person_class_index = classes.index('person')

image_url = mu_df['이미지URL'][:10]

# 저장할 디렉토리 생성
output_dir = '/content/MyDrive/MyDrive/save_png/person'
not_output_dir = '/content/MyDrive/MyDrive/save_png/not_person'

#이미지 저장 경로 설정
a = [output_dir,not_output_dir]
for i in a:
  if not os.path.exists(i):
      os.makedirs(i)
person_img_url_list=[]
not_person_img_url_list=[]

#이미지 로드 함수
def download_image(url):
    try:
      response = requests.get(url)
    except:
      response = requests.get("https:" + url)
    img = Image.open(BytesIO(response.content))
    return cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)
# 이미지로 사람 판별
def detect_person(image):
    height, width = image.shape[:2]
    blob = cv2.dnn.blobFromImage(image, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)
    outs = net.forward(net.getUnconnectedOutLayersNames())

    # 결과 분석
    person_detected = False
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if class_id == person_class_index and confidence > 0.5:  # 신뢰도 임계값 0.5
                person_detected = True
                break
        if person_detected:
            break
    return person_detected

#스타일샷+메인이미지 이미지를 롤링해서 처음 사람이 나오면 저장후 다음롤링으로 넘어감
for index,i in enumerate(image_url):
    image_urls = i.split(',')
    count=0
    for idx, url in enumerate(image_urls):
        print(url)
        if count > 0:
          continue

        image = download_image(url.strip())

        count=0
        if detect_person(image):
            filename = os.path.join(output_dir, f'person_{index}.jpg')
            cv2.imwrite(filename, image)
            person_img_url_list.append(url)
            print(f'Saved: {filename}')
            count+=1



        else:
            filename = os.path.join(not_output_dir, f'not_person_{index}.jpg')
            cv2.imwrite(filename, image)
            not_person_img_url_list.append(url)
            print(f'No person detected in image {index}')

```

```python
import os
import requests
from PIL import Image
from io import BytesIO
#배경제거
def remove_background(img_url):
    res = requests.get(img_url)
    img = Image.open(BytesIO(res.content))
    return img
#배경제거 이미지와 제거 전 이미지를 따로 저장
def download_images(image_urls, folder_path, original_folder_path):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
    if not os.path.exists(original_folder_path):
        os.makedirs(original_folder_path)

    for idx, url in enumerate(image_urls):
        try:
            # 배경 제거 전에 원본 이미지 다운로드 및 저장
            original_img = Image.open(BytesIO(requests.get(url).content))
            original_img.save(os.path.join(original_folder_path, f"original_image_{idx}.jpg"))

            # 배경 제거된 이미지 다운로드 및 저장
            output = remove_background(url)
            output = output.convert("RGB")
            output.save(os.path.join(folder_path, f"image_{idx}.jpg"))
            print(f"이미지 {idx} 다운로드 및 저장 완료")
        except Exception as e:
            print(f"이미지 {idx} 다운로드 및 저장 중 오류 발생: {e}")

# 이미지 URL 리스트
image_urls = [
    'https://image.musinsa.com/images/codimap/list/2024050110255100000014910.jpg',
    'https://image.msscdn.net/mfile_s01/_shopstaff/list.staff_6629c4ea75ad8.jpg',
    'https://image.msscdn.net/mfile_s01/_shopstaff/list.staff_661ceb6d0de07.jpg',
    'https://image.msscdn.net/mfile_s01/_shopstaff/list.staff_6629ee2a9ff0e.jpg',
    'https://image.msscdn.net/mfile_s01/_shopstaff/list.staff_661ce9a4ae465.jpg',
    'https://image.msscdn.net/mfile_s01/_shopstaff/list.staff_661ce8e2749f8.jpg',
    'https://image.msscdn.net/mfile_s01/_shopstaff/list.staff_625cbe8039a2f.jpg',
    'https://image.msscdn.net/mfile_s01/_shopstaff/list.staff_62591e63df2d5.jpg'
]

# 저장할 폴더 경로
folder_path = "/content/save_png"
original_folder_path = "/content/save_original_png"

# 이미지 다운로드 및 저장
download_images(image_urls, folder_path, original_folder_path)

```

```python
#이미지를 직접 할당해서 잘 돌아가는지 확인
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
import torch
import os
import cv2
from yolo.utils.utils import *
from predictors.YOLOv3 import YOLOv3Predictor
import glob
from tqdm import tqdm
import sys
import uuid


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.cuda.empty_cache()
print(device)

#YOLO PARAMS

yolo_modanet_params = {   "model_def" : "yolo/modanetcfg/yolov3-modanet.cfg",
"weights_path" : "yolo/weights/yolov3-modanet_last.weights",
"class_path":"yolo/modanetcfg/modanet.names",
"conf_thres" : 0.5,
"nms_thres" :0.4,
"img_size" : 416,
"device" : device}


#DATASET
# 데이터셋 설정
dataset = 'modanet' #'df2'

#YOLO 파라미터 설정
yolo_params = yolo_modanet_params

#Classes
# 클래스 로드
classes = load_classes(yolo_params["class_path"])

#Colors
# 색상 설정
cmap = plt.get_cmap("rainbow")
colors = np.array([cmap(i) for i in np.linspace(0, 1, 13)])
#np.random.shuffle(colors)

# 모델 설정
model = 'yolo'
# YOLOv3 모델로 객체 검출기 생성
detectron = YOLOv3Predictor(params=yolo_params)



# 저장할 디렉토리 생성
save_dir = "object_images"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)

while(True):
    # 사용자로부터 이미지 경로 입력 받기
    path = input('이미지 경로: ')
    # 이미지가 존재하는지 확인
    if not os.path.exists(path):
        print('이미지가 존재하지 않습니다..')
        continue
    # 이미지 읽어오기
    img = cv2.imread(path)
    # 객체 검출 수행
    detections = detectron.get_detections(img)
    # 검출된 객체가 있을 때
    if len(detections) != 0:
        # 검출된 객체들을 신뢰도에 따라 정렬
        detections.sort(reverse=False, key=lambda x: x[4])
        # 객체 수만큼 반복
        for idx, (x1, y1, x2, y2, cls_conf, cls_pred) in enumerate(detections):
            # 객체 정보 출력
            print("\t+ label: %s, conf: %.5f" % (classes[int(cls_pred)], cls_conf))
            # 색상 설정
            color = colors[int(cls_pred)]
            color = tuple(c*255 for c in color)
            color = (.7*color[2], .7*color[1], .7*color[0])
            # 폰트 설정
            font = cv2.FONT_HERSHEY_SIMPLEX
            # 객체 주위에 박스 그리기
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
            text = "%s conf: %.3f" % (classes[int(cls_pred)], cls_conf)
            cv2.rectangle(img, (x1, y1), (x2, y2), color, 3)
            # 텍스트 위치 설정
            y1 = 0 if y1 < 0 else y1
            y1_rect = y1 - 25
            y1_text = y1 - 5
            if y1_rect < 0:
                y1_rect = y1 + 27
                y1_text = y1 + 20
            # 텍스트와 박스 배경 그리기
            cv2.rectangle(img, (x1-2, y1_rect), (x1 + int(8.5*len(text)), y1), color, -1)
            cv2.putText(img, text, (x1, y1_text), font, 0.5, (255, 255, 255), 1, cv2.LINE_AA)
            # 객체 영역 이미지 저장
            object_img = img[y1:y2, x1:x2]
            cv2.imwrite(os.path.join(save_dir, f"object_{idx}.jpg"), object_img)
    # 이미지 출력
    img_id = path.split('/')[-1].split('.')[0]
    cv2_imshow(img)
    print('내부 루프 종료')
    break
print("외부 루프 종료")
```
