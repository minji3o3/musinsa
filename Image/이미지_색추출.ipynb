```python
def remove_background(img_url):
  res=requests.get(img_url)
  img=Image.open(BytesIO(res.content))
  output=remove(img)
  display(output)

  return output
```

```python
num_clusters=3
```

```python
def extract_color_img(img):
  img_cv=cv2.cvtColor(np.array(img),cv2.COLOR_RGB2BGR)
  img_rgb=cv2.cvtColor(img_cv,cv2.COLOR_BGR2RGB)

  # 검정색을 필터링하여 추출된 이미지에서 픽셀을 가져옴
  pixels = img_rgb[np.any(img_rgb != [0, 0, 0], axis=-1)]

  kmeans=KMeans(n_clusters=num_clusters)
  kmeans.fit(pixels)

  cluster_centers=kmeans.cluster_centers_ #주요색상
  cluster_labels=kmeans.labels_

  unique,cnts=np.unique(cluster_labels,return_counts=True) #클러스터 픽셀 수
  color_cnt=dict(zip(unique,cnts))

  #클러스터 정렬
  sorted_clusters=sorted(color_cnt.items(),key=lambda x:x[1])

  #첫번째 메인 클러스터 색상
  dominant_color_index=max(color_cnt,key=color_cnt.get)
  dominant_color=cluster_centers[dominant_color_index]
  dominant_int=dominant_color.astype(int)
  dominant_str=[str(d) for d in dominant_int]

  #두번째 메인 클러스터 색상
  second_dominant_color_index=sorted_clusters[1][0]
  second_dominant_color=cluster_centers[second_dominant_color_index]
  second_dominant_int=second_dominant_color.astype(int)
  second_dominant_str=[str(d) for d in second_dominant_int]

  # 시각화코드
  colors = [cluster_centers[i] / 255 for i in range(num_clusters)]
  sizes = [color_cnt[i] for i in range(num_clusters)]

  # 비율 계산
  sum_size=sum(sizes)
  ratios=[ int(sizes[i])/sum_size for i in range(num_clusters)]
  ratios.sort(reverse=True)
  # print("r",ratios)

  plt.pie(sizes, labels=[f"Cluster {i}" for i in range(num_clusters)], colors=colors, autopct='%1.1f%%')
  plt.title("Dominant Colors in the Image")
  plt.show()

  # 비율이 0.5이상인 경우엔 메인 컬러1개, 0.5 미만인 경우 2개 추출-> ex. 체크셔츠
  if ratios[0]>=0.5:
    return ",".join(dominant_str)
  else:
    return ",".join(dominant_str+second_dominant_str)
```

```python
def extract_main_color(img_url):
  removed=remove_background(img_url)
  main_color=extract_color_img(removed)

  return main_color
```
