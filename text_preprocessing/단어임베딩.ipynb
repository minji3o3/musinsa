```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/minzy/ASAC/머신러닝프로젝트/전처리/representwords_split.csv', encoding = 'utf-8')
del df['Unnamed: 0']
df.head()
```

```python
df['word1'] = df['word1'].str.replace('+', '')
df['word2'] = df['word2'].str.replace('+', '')
df['word3'] = df['word3'].str.replace('+', '')
```

```python
df['word1'] = df['word1'].str.replace(' ', '')
df['word2'] = df['word2'].str.replace(' ', '')
df['word3'] = df['word3'].str.replace(' ', '')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vector = TfidfVectorizer(norm = None)
tfidf_matrix  = tfidf_vector.fit(df['representative_words'])
vocab = list(tfidf_vector.vocabulary_.keys())
vocab.sort()
```

```python
vocab
```

```python
len(vocab)
```

```python
tfidf_df = pd.DataFrame(tfidf_vector.transform(df['representative_words']).toarray(), columns = vocab)
tfidf_df
```

```python
feature_names = tfidf_matrix.get_feature_names_out()
average_tfidf_scores = tfidf_matrix.transform(df['representative_words']).mean(axis = 0).tolist()[0]
word_scores = dict(zip(feature_names, average_tfidf_scores))
```

```python
for word, score in sorted(word_scores.items(), key = lambda x : x[1], reverse = True) :
    print(f'{word} : {score}')
```

```python
sorted_dict = {word: score for word, score in sorted(word_scores.items(), key=lambda x: x[1], reverse=True)}
```

```python
sorted_dict
```

```python
df.info()
```

```python
df = df[['브랜드', 'representative_words', 'word1', 'word2', 'word3']]
df
```

```python
df['word1_score'] = df['word1'].map(sorted_dict)
df['word2_score'] = df['word2'].map(sorted_dict)
df['word3_score'] = df['word3'].map(sorted_dict)
```

```python
df['represent_words_score'] = df[['word1_score', 'word2_score', 'word3_score']].values.tolist()
```
