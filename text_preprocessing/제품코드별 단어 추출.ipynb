```python
def text_preprocessor(s):

    s = s.replace('니다', '')
    s = s.replace('쁘네', '')
    s = s.replace('들다', '')
    s = s.replace('상이', '')
    s = s.replace('진짜', '')
    s = s.replace('가다', '')
    s = s.replace('고요', '')
    s = s.replace('이기다', '')
    s = s.replace('쁠듯', '')
    s = s.replace('파파팟', '')
    s = s.replace('너무', '')
    s = s.replace('세영', '')
    s = s.replace('색도', '')
    s = s.replace('빨', '')
    s = s.replace('함', '')
    s = s.replace('샴', '')
    s = s.replace('남', '')
    s = s.replace('줌', '')
    s = s.replace('쁠', '')
    s = s.replace('로라', '')
    s = s.replace('누반', '')
    s = s.replace('강', '')
    s = s.replace('반', '')
    s = s.replace('트릴', '')
    s = s.replace('리온', '')
    s = s.replace('가세', '')
    s = s.replace('대요', '')
    s = s.replace('자친구', '')
    s = s.replace('케', '')
    s = s.replace('겟거', '')
    s = s.replace('긱', '')
    s = s.replace('려', '')
    s = s.replace('방', '')
    s = s.replace('컷', '')
    s = s.replace('쁠', '')
    s = s.replace('쁘', '')
    s = s.replace('군데', '')
    s = s.replace('곳', '')
    s = s.replace('닌', '')
    s = s.replace('룩입니', '')
    s = s.replace('세터', '')
    s = s.replace('벅', '')
    s = s.replace('롭', '')
    s = s.replace('벼워', '')
    s = s.replace('고료', '')
    s = s.replace('날', '')
    s = s.replace('블랭', '')
    s = s.replace('크룸', '')
    s = s.replace('꼭', '')
    s = s.replace('금', '')
    s = s.replace('벼워', '')
    s = s.replace('네오', '')
    s = s.replace('쁩니', '')
    s = s.replace('매우', '')
    s = s.replace('삿어', '')
    s = s.replace('컨', '')
    
    pattern = r'[^a-zA-Z가-힣]'
    s = re.sub(pattern = pattern, repl = ' ', string = s)

    return s
```

```python
code_group = df.groupby('제품코드')['pos_nouns'].apply(list).reset_index()
code_group['split_words'] = code_group['pos_nouns'].apply(lambda x : [i.split(' ') for i in x])
code_group['split_words'] = code_group['split_words'].apply(lambda x: ' '.join([i for sublist in x for i in sublist]))
code_group
```

```python
code_group['split_words'] = code_group['split_words'].apply(lambda x: text_preprocessor(x))
code_group['split_words'] = code_group['split_words'].str.split( )
code_group
```

```python
from collections import Counter
```

```python
def sort_word_frequency(data) :

    count = Counter(data)
    sorted_lst = sorted(data, key = lambda x : (-count[x], x))

    no_duplicated = []

    for i in sorted_lst :

        if i not in no_duplicated :
            if i not in no_datas :
                no_duplicated.append(i)

        if len(no_duplicated) == 5 :
            break

    return no_duplicated
```

```python
code_group['representative_words'] = code_group['split_words'].apply(sort_word_frequency)
code_group
```
