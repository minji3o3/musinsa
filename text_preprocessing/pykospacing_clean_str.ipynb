```python
df = df.drop_duplicates()
df.reset_index(drop = True, inplace = True)
df.head()
```

```python
df.columns = ['product_code', 'review_type', 'level', 'id',
                          'date', 'person_spec', 'size', 'score', 'review']
df.head()
```

```python
import re
```

```python
def clean_str(text):
    pattern = '[^\w\s]'
    text = re.sub(pattern = pattern, repl = '', string = text)
    return text
```

```python
df.review = df.review.apply(clean_str)
df.head()
len(df)
```

```python
df['review'] = df['review'].str.replace('+','') # 다중 공백 제거
df.head()
```

```python
from pykospacing import Spacing
```

```python
from tqdm import tqdm
```

```python
spacing = Spacing()
review = []

for i in tqdm(df['review'][500000:600000]) :
  a = spacing(i)
  review.append(a)
```

```python
len(review)
```

```python
review_ = pd.DataFrame(review, columns = ['pre_review'])
review_.to_csv('review_preprocessing(500000-600000).csv', encoding = 'utf-8')
```
