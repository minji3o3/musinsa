```python
def clean(x) :
    
    pattern = r'[^가-힣]'
    x = re.sub(pattern = pattern, repl = ' ', string = x)
    
    return x
```

```python
def replaced(s):

    s = s.replace('색감', '색')
    s = s.replace('색상', '색')
    s = s.replace('커서', '크기')
    
    return s
```

```python
df['word1'] = df['word1'].apply(lambda x: replaced(x))
df['word2'] = df['word2'].apply(lambda x: replaced(x))
df['word3'] = df['word3'].apply(lambda x: replaced(x))
```
